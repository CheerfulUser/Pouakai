```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
%matplotlib notebook

from scipy.ndimage import gaussian_filter
```

```python
files = glob('zp_surface/*R-10_*.npy')
print(len(files))
```

```python
arrs = []
for f in files:
    arrs += [np.load(f)]
arrs = np.array(arrs)
```

```python
a = 10**(-2/5*arrs)
```

```python
med = np.nanmedian(a / np.nanmedian(a,axis=(1,2))[:,np.newaxis,np.newaxis],axis=0)
std = np.nanstd(a / np.nanmedian(a,axis=(1,2))[:,np.newaxis,np.newaxis],axis=0)
```

```python
plt.figure()
plt.imshow(med)
plt.colorbar()
plt.tight_layout()
plt.figure()
plt.imshow(std/med)
plt.colorbar()
plt.tight_layout()
```
